# Split the train dataframe into 4 parts

```python
# Define the number of parts and rows per part
num_parts = 4

# Calculate rows for the first three parts and the last part
rows_per_part = len(train_df) // num_parts
remaining_rows = len(train_df) % num_parts

# Split the DataFrame and save each part
for i in range(num_parts):
    # Calculate start and end indices for the split
    start_index = i * rows_per_part
    if i < remaining_rows:
        # Add 1 extra row for the first 3 parts
        end_index = start_index + rows_per_part + 1
    else:
        # For the last part, no extra rows
        end_index = start_index + rows_per_part

    # Create a sub-DataFrame for the current part
    part_df = train_df.iloc[start_index:end_index]
    
    # Define the file path for saving (adjust the path as needed)
    part_df_path = f"{dataset_common_path}/train_dataset_split_{i + 1}.csv"
    
    # Save the sub-DataFrame to a CSV file
    part_df.to_csv(part_df_path, index=False)
    
    print(f"Shape of train_dataset_split_{i + 1} is: {part_df.shape}")
    print(f"{part_df_path} has been split and saved successfully\n")
```
